# Interactive plots with ECG data

```python
import numpy as np
import scipy as sc
from scipy.io import loadmat
import matplotlib.pylab as plt
import seaborn as sns; sns.set()
sns.set(color_codes=True)
from ipywidgets import widgets
```

```python
%matplotlib inline
```

```python
p1a = loadmat('P1A.mat')
p1c = loadmat('P1C.mat')
p1d = loadmat('P1D.mat')

data_a = p1a['data']
ecg_a = data_a[:,0]
eda_a = data_a[:,1]
resp_a = data_a[:,2]
rr_a = data_a[:,3]
hr_a = data_a[:,4]

data_c = p1c['data']
ecg_c = data_c[:,0]
eda_c = data_c[:,1]
resp_c = data_c[:,2]
rr_c = data_c[:,3]
hr_c = data_c[:,4]

data_d = p1d['data']
ecg_d = data_d[:,0]
eda_d = data_d[:,1]
resp_d = data_d[:,2]
rr_d = data_d[:,3]
hr_d = data_d[:,4]
```

```python
def plot_ecg(x0,xfin):
    plt.figure(figsize=(15,8))
    plt.plot(ecg_a[x0:xfin])
    plt.show()
widgets.interact(plot_ecg, x0=(0,903245,1) , xfin=(0,903245,1))
```

```python
def plot_vals(x0, xfin0, x1, xfin1, x2, xfin2):
    plt.figure(figsize=(15,8),dpi=1800)
    plt.plot(hr_a[x0:xfin0], 'b')
    ma = np.mean(hr_a[x0:xfin0])
    mc = np.mean(hr_c[x1:xfin1])
    md = np.mean(hr_d[x2:xfin2])
    plt.hold
    plt.plot(hr_c[x1:xfin1], 'g')
    plt.plot(hr_d[x2:xfin2], 'r')
    mga = ma*np.ones(np.size(hr_d))
    mgc = mc*np.ones(np.size(hr_d))
    mgd = md*np.ones(np.size(hr_d))
    plt.plot(mga[x0:xfin0], 'b',lw=2)
    plt.plot(mgc[x1:xfin1], 'g',lw=2)
    plt.plot(mgd[x2:xfin2], 'r',lw=2)
    plt.ylabel( 'HR', fontsize=20)
    plt.xlabel('time' +  '\n\n\nMean A: ' + str(round(mga[0],2)) + ' bpm\n'
               + 'Mean C: ' + str(round(mgc[0],2)) + ' bpm\n'
               + 'Mean D:  ' + str(round(mgd[0],2)) + ' bpm\n' , fontsize=24)
    plt.xlim(min(x0,x1,x2), min(xfin0,xfin1,xfin2))
    plt.show()
widgets.interact(plot_vals, x0=(0,1200000,100) , xfin0=(0,1000000,100),
                            x1=(0,1200000,100) , xfin1=(0,1000000,100),
                            x2=(0,1200000,100) , xfin2=(0,1000000,100))
```

```python
def plot_hist(x0, xfin0, x1, xfin1, x2, xfin2):
    plt.figure(figsize=(12, 10), dpi=1600)
    plt.hist(hr_a[x0:xfin0], bins=50, alpha=0.6, color='b')
    plt.axvline(hr_a[x0:xfin0].mean(), color='b', linestyle='dashed', linewidth=3)
    plt.hold
    plt.hist(hr_c[x1:xfin1], bins=50, alpha=0.6, color='g')
    plt.axvline(hr_c[x1:xfin1].mean(), color='g', linestyle='dashed', linewidth=3)
    plt.hist(hr_d[x2:xfin2], bins=50, alpha=0.6, color='r')
    plt.axvline(hr_d[x2:xfin2].mean(), color='r', linestyle='dashed', linewidth=3)
    plt.ylabel('Density', fontsize=24)
    plt.xlabel('HR' +  '\n\n\nMean A: ' + str(round(hr_a[x0:xfin0].mean(),2)) + ' bpm\n'
               + 'Mean C: ' + str(round(hr_c[x1:xfin1].mean(),2)) + ' bpm\n'
               + 'Mean D:  ' + str(round(hr_d[x2:xfin2].mean(),2)) + ' bpm\n' , fontsize=24)
    plt.show()
widgets.interact(plot_hist, x0=(0,1200000,100), xfin0=(0,1000000,100),
                            x1=(0,1200000,100), xfin1=(0,1000000,100),
                            x2=(0,1200000,100), xfin2=(0,1000000,100))
```

### Poincare plots of RR intervals

```python
def plot_scat(xmin,xmax):
    plt.figure(figsize=(12, 10))
    min1 = rr_d[xmin:round((xmin+xmax)/2)]
    min2 = rr_d[round((xmin+xmax)/2):xmax]
    plt.scatter(min1, min2, color='r')
widgets.interact(plot_scat, xmin=(0,900000,100) , xmax=(0,900000,100))
```

```python
def plot_scat(xmin,xmax):
    plt.figure(figsize=(12, 10))
    min1 = rr_c[xmin:round((xmin+xmax)/2)]
    min2 = rr_c[round((xmin+xmax)/2):xmax]
    plt.scatter(min1, min2, color='g')
widgets.interact(plot_scat, xmin=(0,900000,100) , xmax=(0,900000,100))
```
